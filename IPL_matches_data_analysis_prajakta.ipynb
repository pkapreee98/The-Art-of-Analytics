What is the Number of unique Player of the Match in the given dataset ?

```python
unique_player=set(df['Player_of_Match'])
n=len(unique_player)
print('The unique Players of the match are', n)
```

What is the mean, median and standard deviation of 'Margin' in the dataset ?

```python
mean=df['Margin'].mean()
median=df['Margin'].median()
sd=df['Margin'].std()
mean, median, sd
```

Which player has received the most number of Player of the match award

```python
freq=df['Player_of_Match'].value_counts()
freq=freq.idxmax()
print('The player who received max number of Player of the match is', freq)
```

Name the Team which has the highest Margin of Win ?

Make a pie chart to analyse Team1 with mumber of match played? Label the Pie Chart.

```python
# Count the occurrences of each category

count_data = df['Team1'].value_counts().reset_index()
#count_data
plt40.pie(count_data['Team1'],labels=count_data['index'])
```

Bar chart of Umpire1 column and number of match they umpired? Set figure size as (100,20)

```python
# Count the occurrences of each category
plt40.figure(figsize=(100, 20))
count_data = df['Umpire1'].value_counts().reset_index()

plt40.bar(count_data['index'],count_data['Umpire1'])

# Set x-label and y-label
plt40.xlabel('Umpire1 Names',fontsize=70)
plt40.ylabel('No. of matches umpired',fontsize=70)

# Set title
plt40.title('Bar Chart',fontsize=100)
plt40.grid()

```
