```python
# Question number 2:
import numpy as np
import statistics
dataset=np.array([5,3,6,4,3,6,9,0,4,1])
# to find mean
mean=np.mean(dataset)

# to find median
sorted_dataset=np.sort(dataset)
median=np.median(dataset)

# to find mode
mode=statistics.mode(dataset)
mean,median, mode,sorted_dataset

Q1=np.percentile(dataset,25)
Q3=np.percentile(dataset,75)
Q1,Q3,Q3-Q1
```
